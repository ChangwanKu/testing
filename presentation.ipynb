```python
###############로또 생성기
from random import shuffle
import matplotlib

gamenum = input('로또 게임 회수를 입력하세요: ')
testing = []
for i in range(int(gamenum)):
    balls = [x for x in range(1,46)]
    ret = []
    for j in range(6):
        shuffle(balls)
        number = balls.pop()
        ret.append(number)
        testing.append(number)
    ret.sort()
    #print('로또번호[%d]: ' %(i+1), end='')
    #print(ret)
lot = []
#1~45 수 만들기
for i in range(1,46):
    lot.append(i)
y = []
for i in lot:
    y.append(testing.count(i))

%matplotlib inline
#print(testing)
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
plt.bar(lot, y)

plt.title('lotto')
plt.xlabel('num')
plt.ylabel('freq')
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)

```
