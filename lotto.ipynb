```python
#####로또 744회까지 숫자들

import pandas as pd
import os
import matplotlib

lotto = pd.read_csv('lotto.csv',header=None,usecols = [2,3,4,5,6,7])

colnum = list( lotto.columns)
#리스트 하나로 뽑기
l_list = []
for i in colnum:
    l_list += (lotto[i].tolist())
lot = []
#1~45 수 만들기
for i in range(1,46):
    lot.append(i)
    
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
y = []
for i in lot:
    y.append(l_list.count(i))
lot = []
for i in range(1,46):
    lot.append(i)

plt.bar(lot, y)
plt.title('lotto')
plt.xlabel('num')
plt.ylabel('freq')
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)
#fig.savefig('test2png.png', dpi=100)
```
